```python
import pandas as pd
import numpy as np
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk
nltk.download('punkt')
nltk.download('stopwords')
import matplotlib.pyplot as plt
import spacy
from textblob import TextBlob
import re
#import ssl
#from collections import Counter
```

```python
desc = pd.read_csv('monster_desc.csv')
gender = pd.read_csv('monster_gen.csv')
```

```python
# Exclude titles and non-monsters 
desc = desc[desc['key'].str.contains("descriptionKey")]
desc = desc[desc['key'].str.contains("monster")]
```

```python
desc.head()
```

```python
def get_character_name(string):
    new_string = string.str.lower().str.replace('_', ' ').str.split().str[1]
    return new_string

def clean_and_tokenize_description(df):
    d = (df
    .assign(desc_clean= lambda x: x['dev_en'].str.lower()
    ))
    
    list_desc = list(d['desc_clean'])
    text_desc = ''.join(list_desc)
    td = re.sub(r'[^\w\s]','', text_desc)
    
    tm_tokens = word_tokenize(td)
    return tm_tokens

def extract_adjectives(text):
    blob = TextBlob(text)
    return [ word for (word,tag) in blob.tags if tag == "JJ"]
```

```python
#Merge descriptions with gender
d = (desc
     .assign(
    name=lambda x: get_character_name(x['key'])
    )
)
g = (gender
     .assign(
    name= lambda x: x['Name'].str.lower(),
    Gender= lambda x: x['Gender'].str.lower())
)
df = pd.merge(d,g, on = 'name')
```

```python
male = (df
       .query("Gender == 'male'"))
female = (df
       .query("Gender == 'female'"))
neutral = (df
       .query("Gender == 'neutral'"))
```

```python
#male.head()
```

```python
male_tokens = clean_and_tokenize_description(male)
female_tokens = clean_and_tokenize_description(female)
neutral_tokens = clean_and_tokenize_description(neutral)
```

```python
tm_tokens_without_sw = [word for word in male_tokens if not word in stopwords.words()]
#tm_without_sw = ' '.join(tm_tokens_without_sw)
#m_adjectives = get_adjectives(tm_without_sw)
#m_adj = ' '.join(m_adjectives)
```

```python
tf_tokens_without_sw = [word for word in female_tokens if not word in stopwords.words()]
tf_without_sw = ' '.join(tf_tokens_without_sw)
f_adjectives = extract_adjectives(tf_without_sw)
f_adj = ' '.join(f_adjectives)
```

```python
tn_tokens_without_sw = [word for word in neutral_tokens if not word in stopwords.words()]
tn_without_sw = ' '.join(tn_tokens_without_sw)
n_adjectives = extract_adjectives(tn_without_sw)
n_adj = ' '.join(n_adjectives)
```

```python
wordcloud = WordCloud(stopwords=STOPWORDS).generate(m_adj)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wordcloud = WordCloud(stopwords=STOPWORDS).generate(f_adj)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wordcloud = WordCloud(stopwords=STOPWORDS).generate(n_adj)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
common_words = ['new', 'little', 'many', 'good', 'great', 'isnt', 'doesnt', 'dr', 'become', 'last', 'first'
                 'sure', 'magic', 'light', 'much', 'become',  'present', 'ready', 'true', 'first', 'live']
m_adjectives_wo_common_words = [word for word in m_adjectives if not word in common_words]
m_adj_wocw = ' '.join(m_adjectives_wo_common_words)
f_adjectives_wo_common_words = [word for word in f_adjectives if not word in common_words]
f_adj_wocw = ' '.join(f_adjectives_wo_common_words)
n_adjectives_wo_common_words = [word for word in n_adjectives if not word in common_words]
n_adj_wocw = ' '.join(n_adjectives_wo_common_words)
```

```python
wordcloud = WordCloud(stopwords=STOPWORDS).generate(m_adj_wocw)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
